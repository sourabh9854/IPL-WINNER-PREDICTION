```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from pandas import DataFrame
import pickle
```

```python
matches= pd.read_csv(r"IPL_Matches_2008_2022[1].csv")
matches.head()
```

```python
delivery = pd.read_csv(r"IPL_Ball_by_Ball_2008_2022[1].csv")
delivery.head()
```

```python
total_score =delivery.groupby(['ID','innings']).sum()['total_run'].reset_index()
```

```python
total_score=total_score[total_score["innings"]==1]
```

```python
match_df=pd.merge(matches,total_score,left_on=['ID'], right_on = ['ID']) 
```

```python
match_df.drop(['innings'],axis=1)
match_df.head()
match_df.shape
```

```python
match_df['Team1'] = match_df['Team1'].replace(['Deccan Chargers'], 'Sunrisers Hyderabad')
match_df['Team2'] = match_df['Team2'].replace(['Deccan Chargers'], 'Sunrisers Hyderabad')
match_df['TossWinner'] = match_df['TossWinner'].replace(['Deccan Chargers'], 'Sunrisers Hyderabad')
match_df['WinningTeam'] = match_df['WinningTeam'].replace(['Deccan Chargers'], 'Sunrisers Hyderabad')


match_df['Team1'] = match_df['Team1'].replace(['Delhi Daredevils'], 'Delhi Capitals')
match_df['Team2'] = match_df['Team2'].replace(['Delhi Daredevils'], 'Delhi Capitals')
match_df['TossWinner'] = match_df['TossWinner'].replace(['Delhi Daredevils'], 'Delhi Capitals')
match_df['WinningTeam'] = match_df['WinningTeam'].replace(['Delhi Daredevils'], 'Delhi Capitals')



match_df['Team1'] = match_df['Team1'].replace(['Kings XI Punjab'], 'Punjab Kings')
match_df['Team2'] = match_df['Team2'].replace(['Kings XI Punjab'], 'Punjab Kings')
match_df['TossWinner'] = match_df['TossWinner'].replace(['Kings XI Punjab'], 'Punjab Kings')
match_df['WinningTeam'] = match_df['WinningTeam'].replace(['Kings XI Punjab'], 'Punjab Kings')
```

```python
match_df['Team1'].unique()
```

```python
teams = ['Rajasthan Royals', 'Royal Challengers Bangalore',
       'Sunrisers Hyderabad', 'Delhi Capitals', 'Chennai Super Kings',
       'Gujarat Titans', 'Lucknow Super Giants', 'Kolkata Knight Riders',
       'Punjab Kings', 'Mumbai Indians' ]
```

```python
match_df =match_df[match_df['Team1'].isin(teams)]
match_df =match_df[match_df['Team2'].isin(teams)]
match_df =match_df[match_df['TossWinner'].isin(teams)]
match_df =match_df[match_df['WinningTeam'].isin(teams)]
```

```python
match_df['Team2'].unique()
```

```python
match_df.isnull().sum()
```

```python
match_df['method'].fillna('0', inplace=True)
```

```python
meth = ['0']
```

```python
match_df =match_df[match_df['method'].isin(meth)]
```

```python
match_df=match_df[['ID','City','WinningTeam','total_run','Team1','Team2']]
```

```python
ipl_df=pd.merge(delivery,match_df,left_on=['ID'], right_on = ['ID'])
ipl_df['BattingTeam'] =ipl_df['BattingTeam'].replace(['Kings XI Punjab'], 'Punjab Kings')
ipl_df['BattingTeam'] =ipl_df['BattingTeam'].replace(['Delhi Daredevils'], 'Delhi Capitals')
ipl_df['BattingTeam'] = ipl_df['BattingTeam'].replace(['Deccan Chargers'], 'Sunrisers Hyderabad')
```

```python
ipl_df.loc[ipl_df["BattingTeam"]==ipl_df["Team1"],"BowlingTeam"]=ipl_df['Team2']
ipl_df.loc[ipl_df["BattingTeam"]==ipl_df["Team2"],"BowlingTeam"]=ipl_df['Team1']
ipl_df.tail()
```

```python
ipl_df=ipl_df[ipl_df['innings']==2]
```

```python
ipl_df.rename(columns = {'total_run_y':'Target'}, inplace = True)
```

```python
ipl_df.head(12)
```

```python
ipl_df['curr_score']=ipl_df.groupby('ID').cumsum()['total_run_x']
```

```python
ipl_df.head(12)
```

```python
ipl_df['runs_left']=(ipl_df['Target'] - ipl_df['curr_score']) +1
ipl_df
```

```python
ipl_df['ball_left']=(126 - ((ipl_df['overs']+1)*6 +ipl_df['ballnumber']))
```

```python
ipl_df['curr_RR']= ipl_df['curr_score']*6/(120 - ipl_df['ball_left'])
```

```python
 ipl_df['Req_RR']= ipl_df['runs_left']*6/( ipl_df['ball_left'])
```

```python
ipl_df.isnull().sum()
```

```python
ipl_df['wickets_remaining']=10 -ipl_df.groupby('ID').cumsum()['isWicketDelivery']
ipl_df.head()
```

```python
def result(row):
    return 1 if row['BattingTeam'] == row['WinningTeam'] else  0
```

```python
ipl_df['results']=ipl_df.apply(result,axis=1)
```

```python
ipl=ipl_df[['BattingTeam','BowlingTeam','City','Target','runs_left','ball_left','wickets_remaining','curr_RR','Req_RR','results']]
ipl.head()
```

```python
ipl=ipl.sample(ipl.shape[0])
```

```python
ipl.dropna(inplace=True)
```

```python
ipl.describe()
```

```python
ipl = ipl[ipl['ball_left']!=0]
ipl.describe()
```

```python
x = ipl.iloc[:,:-1]
y = ipl.iloc[:,-1]
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.8,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([('trf',OneHotEncoder(sparse=False,drop='first'),['BattingTeam','BowlingTeam','City'])],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
ipl.isnull().sum()
```

```python
pipe = Pipeline(steps=[('step1',trf),('step2',LogisticRegression(solver='liblinear'))])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(x_test)[25]
```

```python
pickle.dump(pipe,open('ipl.pkl','wb'))
```
